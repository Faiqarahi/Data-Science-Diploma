```python
#Covariance heat map of tips data
sns.heatmap(diamonds.corr(),vmin=-1, vmax=1, cmap='carat')
```

```python
sns.distplot(a=diamonds[diamonds['carat']=='C']['price'],bins=3,kde=False)
plt.title("cut")
plt.xticks([0,1])
plt.show()
plt.title("calrity")
sns.distplot(a=diamonds[diamonds['carat']=='Q']['price'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
plt.title("color")
sns.distplot(a=diamonds[diamonds['carat']=='S']['price'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
#Most of the Passengers embarked from Southampton
```

```python
#Plotting correlation by using heatmap
sns.heatmap(diamonds.corr(),cmap='price')
plt.legend()
```

```python
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
diamonds.groupby('cut')['price'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
diamonds.groupby('clairty')['price'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
diamonds.groupby('depth')['price'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
diamonds.groupby('color')['price'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
diamonds.groupby('carat')['price'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="table", y="price", diamonds=diamonds,ax=axesbi[1,1])
sns.boxplot(x="table", y="carat", diamonds=diamonds,ax=axesbi[1,2])
```
